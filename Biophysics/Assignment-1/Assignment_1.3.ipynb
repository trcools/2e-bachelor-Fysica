#### Other tipping points

we have time-dependent fashion tipping points, reversible tipping points and rate-dependent tipping points.

"Strongly nonlinear but reversible transitions are expected to resemble bifurcation-type behaviour so may carry similar early warnings, including slowing down."


There are limitations to the early warnings of bifurcation-type climate tipping points (False alarms --> false positives and missed alarms --> false negatives) 

- False alarms:

  False alarms can arise because signals interpreted as indicative of approaching bifurcation are not statistically robust or have other causes

- Missed alarms:

  Missed alarms can occur if abrupt transitions happen without underlying bifurcation but they can also occur even when bifurcation is approaching.
