# 2.2 Golven

Als voorbeeld:
- $\Psi_1= 5 \cos (2 x - 5t)$
- $\Psi_2= 3 \cos (x - 2t)$
- $\Psi = \Psi_1+ \Psi_2$ als waarschijnlijkheidsamplitude
- $\left| \Psi_1 \right| ^2 = \Psi_1^{\star} \Psi_1 $
- $\left| \Psi_2 \right| ^2 = \Psi_2^{\star} \Psi_2 $
- $\left| \Psi \right| ^2 = \left| \Psi_1 + \Psi _2 \right| ^2 \ne  \left| \Psi_1 \right| ^2 + \left| \Psi_2 \right| ^2$ als positiewaarschijnlijkheidsdichtheid $P(x,t)$

Hieronder eerst illustratie samenstelling van golven; daarna positiewaarschijnlijkheidsdichtheid *(interferentieverschijnselen is wat de kwantummechanica voor deeltjes onderscheidt van de klassieke fysica voor deeltjes)*

# A.2 Dirac delta functie

Uitleg over de Dirac delta functie te vinden in Appendix A (p777-779) van het handboek van Bransden en Joachain

$ \delta \left( x -x^ {\prime} \right) $ kan gezien worden als de limiet van een op 1  genormeerde distributie die scherp gepiekt is bij $x=x^{\prime}$ en waarvoor de breedte naar 0 gaat

Bijvoorbeeld:


1.   $ \delta (x) = \lim _{ \epsilon \to 0^{+}} \frac {\sin \frac {x} {\epsilon}} {\pi x} $ (dit is vergelijking A.23)

2.   $\delta (x) = \lim _{ \epsilon \to 0^{+}} \frac{\epsilon}{\pi} \frac{1} {x^2 + \epsilon ^2}$ (dit is vergelijking A.20)


Met de hierna volgende figuur wordt aangetoond dat voor kleine waarden van $\epsilon$ men een scherp gepiekte functie krijgt

```python
def DiracPres(epsilon,x):
  f1=np.sin(x/epsilon)
  f2=x*np.pi
  return np.divide(f1,f2)

def DiracPres2(epsilon,x):
  f1=np.divide(1.,(np.power(x,2.)+np.power(epsilon,2.)))
  return f1*epsilon/np.pi

# create the plot for a number of epsilon values
# epsilist=[0.001,0.01,0.02,0.1] # illustrates the sharp rise as \epsilon becomes smaller
epsilist=[0.01,0.02,0.1]
colors=['mediumslateblue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan','khaki','fuchsia','gold','tomato']
xval = np.arange(-0.5,0.5+0.005,0.005,dtype=float) # range of x-values
figa=plt.figure(figsize=(24,6))
axa=figa.add_subplot(121)
axb=figa.add_subplot(122)
#
axa.set_title(r'Dirac $\delta(x)$ functie als distributie ',fontsize=20,
              fontweight='bold',color="tomato")
axb.set_title(r'Dirac $\delta(x)$ functie als distributie ',fontsize=20,
              fontweight='bold',color="tomato")
axa.set_xlabel(r'$x$ waarden',fontsize=20)
axb.set_xlabel(r'$x$ waarden',fontsize=20)
axa.set_ylabel(r'$ \left( \sin \frac {x} {\epsilon} \right)  / (\pi x)$',fontsize=20)
axb.set_ylabel(r'$ \epsilon  / \left[ \pi (x^2 + \epsilon ^2) \right]$',fontsize=20)
for ii in range(len(epsilist)):
  axa.plot(xval,DiracPres(epsilist[ii],xval),color=colors[ii],label=r'$\epsilon$={}'.format(epsilist[ii]))
  axb.plot(xval,DiracPres2(epsilist[ii],xval),color=colors[ii],label=r'$\epsilon$={}'.format(epsilist[ii]))
axa.legend(frameon=False,fontsize=17,ncol=1,loc=0) # loc =0: program seeks good location for legend
axb.legend(frameon=False,fontsize=17,ncol=1,loc=0) # loc =0: program seeks good location for legend
```

# 2.4 Golfpaketten

Golfpakketten: lokalisatie d.m.v. sommen van cosinus golven

- Python routine that illustrates how wave packets can be generated by adding together cosine waves
- Illustrates the use of numpy and the creation of simple xy plots


```python
# function that creates a list of wave numbers
def create_list_of_ks(kbegin,kend,kaantal):
    klist=np.linspace(kbegin,kend,kaantal,endpoint=True)
    return klist
#
# function that adds cosine waves with different wave numbers
def create_the_wave(xvalues, klist):
    wave=np.zeros_like(xvalues)
    for k in klist:
        wave+=np.cos(xvalues*k)
    return wave
# MAIN PROGRAM
# STEP1: create the arrays with k numbers
k2=create_list_of_ks(5.,5.25,2) # array with two k numbers
k5=create_list_of_ks(4.75,5.25,5) # array with five k numbers
k7=create_list_of_ks(4.8125,5.1875,7) # array with seven k numbers
k201=create_list_of_ks(4.80,5.20,201) # array with 201 k numbers
#
# an array with x values
number_of_xvalues=2001
xvalues=np.linspace(-130.,+130.,number_of_xvalues,endpoint=True)
#
# STEP2: create the functions by adding cosine waves
wave2=create_the_wave(xvalues, k2)
wave5=create_the_wave(xvalues, k5)
wave7=create_the_wave(xvalues, k7)
wave201=create_the_wave(xvalues, k201)
#
# create the plots in matplotlib
#
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(111)
#
ax.axis([-130.,+130,-2.,14.])
ax.set_title(r'Golfpakketten door $\sum _{i=1}^{i=N}\cos k_i x$',fontsize=20,fontweight='bold',color="Coral")
ax.set_xlabel(r'$x$ waarden')
ax.set_ylabel(r'${\sum _{i=1}^{i=N}\cos k_i x}$ voor verschillende $N$')
ax.plot(xvalues, wave201/201., 'r-', xvalues, wave7/7.+3., 'b-', xvalues, wave5/5.+6., 'g-',xvalues,wave2/2.+9.,'k-')
ax.text(25., -1.,' 201 k waarden',color='red')
ax.text(25., +2.,' 7 k waarden',color='blue')
ax.text(25., +4.5,' 5 k waarden',color='green')
ax.text(25., +7.5,' 2 k waarden',color='black')
ax.text(-75.,11.5,r'Golfpakketten door ${\sum _{i=1}^{i=N}\cos k_i x}$', fontweight='bold')
fig.savefig('AddingCosineWaves.eps')
plt.show()
```

# 2.4 Gaussische golfpakketten

Definitie van een golfpakket:

$
\Psi ( x, t) =  { \frac { 1 } {\sqrt{ 2 \pi \hbar}}
\int d p_x \; \;  }
\exp \left[ \frac {i} {\hbar} \left[ p_x x - E (p_x)
    t \right] \right] {\phi (p_x )} \; \; \; \text{met,} \; \; E (p_x) = \frac{p_x^2} {2m}
$

In analogie met de contour plots voor een golf (golf als functie van $x$ en $t$: zie hierboven) kunnen we het golfpakket visueel voorstellen als functie van $x$ en $t$

Dit doen we voor een Gaussisch golfpakket:
$ \phi ( p_x ) = \frac {1} { \sqrt{ \left( \Delta p_x \right)
    \sqrt{\pi}} } \exp \left[ - \frac {\left( p_x -p_0 \right)^2}
{2 \left( \Delta p_x \right)^2 } \right]
$ (functie phipx)

Alle numerieke berekeningen gebruiken de variabele: $k_x = \frac{p_x} {\hbar}$ zodat $\frac{p_x x} {\hbar} = k_x x$

```python
def phipx(p0,deltapx,pxran):
  normfac=1./(np.sqrt(deltapx*np.sqrt(np.pi)))
  return normfac*np.exp(-np.power((pxran-p0)/(np.sqrt(2.)*deltapx),2.))


def compwavAndPlot(p0,deltapx,pstep,xran):
  pxran=np.arange(p0-3.*deltapx,p0+3.*deltapx,pstep,dtype=float) # range of px-values
  funcphipx=phipx(p0,deltapx,pxran)
  checknorm=np.trapz(np.multiply(funcphipx,funcphipx),pxran) # numpy trapz: numerical integrals
  print(r'CHECK: De norm $\int dp_x \mid \phi(p_x) \mid ^2$ is:{} (zo dicht mogelijk bij 1)'.format(checknorm))

# reeel en imaginair gedeeltje van de golffunctie \Psi(x,t) bij t=0

  realpsi=np.zeros_like(xran,dtype=float) # reeel gedeelte golffunctie
  imagpsi=np.zeros_like(xran,dtype=float) # imaginair gedeelte golffunctie
  psipsi=np.zeros_like(xran,dtype=float)  # gekwadrateerde golffunctie
  sqpi=np.sqrt(2.*np.pi) # wortel uit 2* pi

  for ii in range(len(xran)):
    xx=xran[ii]
    realpsi[ii]=np.trapz(np.multiply(np.cos(xx*pxran),funcphipx),pxran)/sqpi # integraal doen voor elke x :)
    imagpsi[ii]=np.trapz(np.multiply(np.sin(xx*pxran),funcphipx),pxran)/sqpi # integraal doen voor elke x :)

  psipsi=np.add(np.power(realpsi,2.),np.power(imagpsi,2.))
  # numpy.trapz om integralen numeriek te berekenen: https://numpy.org/doc/stable/reference/generated/numpy.trapz.html
  normofpsipsi=np.trapz(psipsi,xran)
  print(r'CHECK: De norm $\int dx \mid \Psi(x,t=0) \mid ^2$ is:{} (zo dicht mogelijk bij 1)'.format(normofpsipsi))
#
# create the plot in matplotlib
#
  fig3,(ax3,ax4)=plt.subplots(1,2,figsize=(24,6))
  ax3.set_title(r'Golffunctie $\Psi(x,t=0) = \frac{1} {\sqrt{2 \pi \hbar}} \int d p_x \exp \left( \frac {i} {\hbar} p_x x \right) \phi (p_x)$',fontsize=20,fontweight='bold',color="Coral")
  ax3.set_xlabel(r'$x$ waarden',fontsize=20)
  ax3.set_ylabel(r'$Re \Psi (x,t=0), Im \Psi (x,t=0) $',fontsize=20)
  ax3.plot(xran,realpsi,label=r'$Re \Psi (x,t=0)$',linestyle='dashed')
  ax3.plot(xran,imagpsi,label=r'$Im \Psi (x,t=0)$',linestyle='dashdot')
  ax3.plot(xran,psipsi,label=r'$ \mid \Psi (x,t=0) \mid ^2$',linestyle='solid')
  ax3.legend(frameon=False,fontsize=17,ncol=1,loc=0) # loc =0: program seeks good location for legend
#
  ax4.set_title(r'Golffunctie $\phi( p_x) = \frac {1} { \sqrt{ \left( \Delta p_x \right) \sqrt{\pi}} } \exp \left[ - \frac {\left( p_x -p_0 \right)^2} {2 \left( \Delta p_x \right)^2 } \right]$',fontsize=20,fontweight='bold',color="Coral")
  ax4.set_xlabel(r'$ k_x = p_x / \hbar$ waarden',fontsize=20)
  ax4.set_ylabel(r'$\phi(p_x / \hbar )$',fontsize=20)
  ax4.plot(pxran,funcphipx,linestyle='solid',label=r'$p_0 / \hbar$={}; $\Delta p_x / \hbar $ ={} '.format(p0,deltapx))
  ax4.legend(frameon=False,fontsize=17,ncol=1,loc=0) # loc =0: program seeks good location for legend
  return()
```

```python
p0=4. ; deltapx=2.0 ; pstep=0.01 # keuze(pstep is the grid size in k_x)
xran=np.arange(-1.8,+1.8,0.1,dtype=float) # een lijst met x waarden waarvoor golffunctie zal berekend worden
compwavAndPlot(p0,deltapx,pstep,xran)
```
