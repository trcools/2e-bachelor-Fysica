In order for the aquarium to fit in the place where you want it, it can be maximally $73$ cm wide ($a$), $60$ cm high ($b$) and $60$ cm deep ($c$). 
Unfortunately, these dimensions are not mentioned by the vendor. 
The only information that you get is: 
- the total volume of the aquarium is $140$ L
- the diagonal of the front plane (line from $A$ to $C$) is $80.6$ cm long 
- the overall diagonal (line from $A$ to $F$) is $94.9$ cm long.

Will the aquarium fit? Define and solve the system of non-linear equations.

### Part 1
Turn the problem at hand into a system of non-linear equations. Write your answer in the following Markdown cell.

$$\mathbf{f}(\mathbf{x})=\begin{bmatrix}
abc\\
a^2+b^2\\
a^2+b^2+c^2
\end{bmatrix}
=
\begin{bmatrix}
140000\\
80.6^2\\
94.9^2
\end{bmatrix}
$$

### Part 2
Solve this system of non-linear equations. Use a reasonable initial guess for the unknown variables $a$, $b$ and $c$. Will the aquarium fit?

```python
# Implement your solution here
```

```python
def f(x):
    return [
        x[0] * x[1] * x[2] - 140000,
        x[0] ** 2 + x[1] ** 2 - 80.6**2,
        x[0] ** 2 + x[1] ** 2 + x[2] ** 2 - 94.9**2,
    ]


sol1 = optimize.root(f, [70, 60, 60])
print(sol1.x)
```
